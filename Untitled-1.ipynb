```python
from tvDatafeed import TvDatafeed, Interval
import pandas as pd 
# Initialize TVDataFeed (make sure you have your TradingView credentials set up)
tv = TvDatafeed()
# Fetch historical data for Nifty 50 on NSE exchange
#BANKNIFTY 27 JAN 2026 CALL 58400  ==> BANKNIFTY260127C58400
data = tv.get_hist(symbol="BANKNIFTY260127C58400", exchange="NSE", interval=Interval.in_1_minute, n_bars=100,fut_contract=None)
# Convert the data to a pandas DataFrame for easier manipulation    
df = pd.DataFrame(data)
# Display the first few rows of the DataFrame       
print(df.head())

#Now fetching data for NIFTY CALL OPTION
#NIFTY 26 JAN 2026 CALL 25450  ==> NIFTY260120C25450
data = tv.get_hist(symbol="NIFTY260120C25450", exchange="NSE", interval=Interval.in_1_minute, n_bars=100,fut_contract=None)
# Convert the data to a pandas DataFrame for easier manipulation    
df = pd.DataFrame(data)
# Display the first few rows of the DataFrame       
print(df.head())

#Now fetching data for NIFTY SPOT
data = tv.get_hist(symbol="NIFTY", exchange="NSE", interval=Interval.in_1_minute, n_bars=100,fut_contract=None)
# Convert the data to a pandas DataFrame for easier manipulation    
df = pd.DataFrame(data)
# Display the first few rows of the DataFrame       
print(df.head())


#Now fetching data for BANKNIFTY SPOT
data = tv.get_hist(symbol="BANKNIFTY", exchange="NSE", interval=Interval.in_1_minute, n_bars=100,fut_contract=None)
# Convert the data to a pandas DataFrame for easier manipulation    
df = pd.DataFrame(data)
# Display the first few rows of the DataFrame       
print(df.head())    


#Feteching NIFTY FUTURE DATA
#fut_contract : int, optional
#None for cash, 1 for continuous current contract in front, 2 for continuous next contract in front . Defaults to None.


data = tv.get_hist(symbol="NIFTY", exchange="NSE", interval=Interval.in_1_minute, n_bars=100,fut_contract=1) #1 means current month future
# Convert the data to a pandas DataFrame for easier manipulation
```
